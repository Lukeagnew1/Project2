```python
import requests 
import pandas as pd
import numpy as np
```

```python
# Performance is v141

headers = {'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'}
   
def get_screener(page):
    screen = requests.get(f'https://finviz.com/screener.ashx?v=141&f=idx_sp500&ft=4&r={page}', headers = headers).text


    tables = pd.read_html(screen)
    tables = tables[-2]
    tables.columns = tables.iloc[0]
    tables = tables[1:]
    
    return tables
```

```python
#Import all the companies in the s and p 500 from finviz 

spy1 = get_screener('0')
spy2 = get_screener('21')
spy3 = get_screener('41')
spy4 = get_screener('61')
spy5 = get_screener('81')
spy6 = get_screener('101')
spy7 = get_screener('121')
spy8 = get_screener('141')
spy9 = get_screener('161')
spy10 = get_screener('181')
spy11 = get_screener('201')
spy12 = get_screener('221')
spy13 = get_screener('241')
spy14 = get_screener('261')
spy15 = get_screener('281')
spy16 = get_screener('301')
spy17 = get_screener('321')
spy18 = get_screener('341')
spy19 = get_screener('361')
spy20 = get_screener('381')
spy21 = get_screener('401')
spy22 = get_screener('421')
spy23 = get_screener('441')
spy24 = get_screener('461')
spy25 = get_screener('481')
spy26 = get_screener('501')
```

```python
# Concate all the seperate pages into one dataframe 

spy = pd.DataFrame()

spy = pd.concat([spy1, spy2, spy3, spy4, spy5, spy6, spy7, spy8, spy9, spy10, spy11, spy12, spy13, spy14, spy15, spy16, spy17, spy18, spy19, spy20, spy21, spy22, spy23, spy24, spy25, spy26], axis=0)
spy = spy.dropna()
spy
```

```python
# Determine a composite score of the 4 performance metrics by finding the mean 

comp_score = pd.DataFrame()

comp_score['Perf Month'] = spy['Perf Month'] 
comp_score['Perf Quart'] = spy['Perf Quart']
comp_score['Perf Half'] = spy['Perf Half']
comp_score['Perf Year'] = spy['Perf Year']
comp_score['Comp Score'] = spy.mean(axis=1)
# comp_score.to_csv(r'C:\Users\Luke\Desktop\Personal\comp.csv')
```

```python
scores = pd.DataFrame()

scores['Ticker'] = spy['Ticker']
scores['Comp Score'] = comp_score['Comp Score']
scores
```
