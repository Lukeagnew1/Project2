```python
import requests 
import pandas as pd
import numpy as np
```

```python
# Performance is v141

headers = {'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'}
   
def get_screener(page):
    screen = requests.get(f'https://finviz.com/screener.ashx?v=141&f=exch_nasd&t=AAPL,ADBE,ADI,ADP,ADSK,AEP,ALGN,AMAT,AMD,AMGN,AMZN,ANSS,ASML,ATVI,AVGO,BIDU,BIIB,BKNG,CDNS,CDW,CERN,CHKP,CHTR,CMCSA,COST,CPRT,CSCO,CSX,CTAS,CTSH,DLTR,DOCU,DXCM,EA,EBAY,EXC,FAST,FB,FISV,FOX,FOXA,GILD,GOOG,GOOGL,HON,IDXX,ILMN,INCY,INTC,INTU,ISRG,JD,KDP,KHC,KLAC,LRCX,LULU,MAR,MCHP,MDLZ,MELI,MNST,MRNA,MRVL,MSFT,MTCH,MU,MXIM,NFLX,NTES,NVDA,NXPI,OKTA,ORLY,PAYX,PCAR,PDD,PEP,PTON,PYPL,QCOM,REGN,ROST,SBUX,SGEN,SIRI,SNPS,SPLK,SWKS,TCOM,TEAM,TMUS,TSLA,TXN,VRSK,VRSN,VRTX,WBA,WDAY,XEL,XLNX,ZM&r={page}', headers = headers).text


    tables = pd.read_html(screen)
    tables = tables[-2]
    tables.columns = tables.iloc[0]
    tables = tables[1:]
    
    return tables
```

```python
qqq1 = get_screener('0')
qqq2 = get_screener('21')
qqq3 = get_screener('41')
qqq4 = get_screener('61')
qqq5 = get_screener('81')
qqq6 = get_screener('101')
```

```python
qqq = pd.DataFrame()

qqq = pd.concat([qqq1, qqq2, qqq3, qqq4, qqq5, qqq6], axis=0)
qqq = qqq.dropna()
qqq
```

```python
qq = pd.DataFrame()

qq['Perf Month'] = qqq['Perf Month']
qq['Perf Quart'] = qqq['Perf Quart']
qq['Perf Half'] = qqq['Perf Half']
qq['Perf Year'] = qqq['Perf Year']
qq
```

```python
qq['Comp Score'] = qq.mean(axis=1)
```

```python
scores = pd.DataFrame()

scores['Ticker'] = qqq['Ticker']
scores['Comp Score'] = qq['Comp Score']
scores
```
